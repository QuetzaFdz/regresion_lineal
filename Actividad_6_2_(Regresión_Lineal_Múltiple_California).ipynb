```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from turtle import color
from sklearn.linear_model import LinearRegression
model= LinearRegression()
```

```python
cal=pd.read_csv('California_limpio.csv')
cal.head()
```

# **host_acceptance_rate vs host_response_rate, Price, number_of_reviews , review_scores_rating**

```python
#Filtramos un tipo de residencia
eha1=cal[cal['room_type'] == 'Entire home/apt']
#Analizamos la dispersión de los datos indpendentes con una gráfica de diseprción
sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color="blue", data=eha1)
sns.scatterplot(x='price', y='host_acceptance_rate', color="red", data=eha1)
sns.scatterplot(x='number_of_reviews', y='host_acceptance_rate', color="red", data=eha1)
sns.scatterplot(x='review_scores_rating', y='host_acceptance_rate', color="yellow", data=eha1)
plt.title("\n Dispersión de variables independientes")
#Declaramos las variables dependientes e independientes
Vars_Indep= eha1[['host_response_rate', 'price','number_of_reviews','review_scores_rating']]
Var_Dep= eha1['host_acceptance_rate']
#Ajustamos el modelo con las variables antes declaradas
model.fit(X=Vars_Indep, y=Var_Dep)
#Verificamos los coeficientes obtenidos para el modelo ajustado
print('\n Modelo matemático: y= 5.67044556e-01x + -1.04428996e-05x + 3.90964494e-04x + 1.95000605e-02x + 0.20318604440746824')
model.__dict__
#Predecimos los valores del nivel de aceptación del host
y_pred= model.predict(X=eha1[['host_response_rate', 'price', 'number_of_reviews', 'review_scores_rating']])
#Insertamos la columna de predicciones
print('\n Añadimos columna de predicciones')
eha1.insert(0, 'Predicciones', y_pred)
display(eha1)
#Coeficiente de determinación
print('\n Coeficiente de determinación')
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
display(coef_Deter)
#Coeficiente de correlación
print('\n Coeficiente de correlación')
coef_Correl=np.sqrt(coef_Deter)
display(coef_Correl)
```
