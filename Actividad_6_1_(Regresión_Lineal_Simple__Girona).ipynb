```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from turtle import color
from sklearn.linear_model import LinearRegression
```

```python
mex=pd.read_csv('California_limpio.csv')
```

# **host_acceptance_rate vs host_response_rate**

```python
eha1=mex[mex['room_type'] == 'Entire home/apt']
display(sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color="blue", data=eha1))
Vars_Indep= eha1[['host_response_rate']]
Var_Dep= eha1['host_acceptance_rate']
model= LinearRegression()
type(model)
model.fit(X=Vars_Indep, y=Var_Dep)
print('\n Predicción de modelo')
display(model.__dict__)
print('\n Eficiencia del modelo')
display(model.score(Vars_Indep,Var_Dep))
print('\n Predecimos valores')
y_pred= model.predict(X=eha1[['host_response_rate']])
display(y_pred)
print('\n Agregamos predicciones')
eha1.insert(0, 'Predicciones', y_pred)
display(eha1)
print('\n Coeficiente de determinación')
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
display(coef_Deter)
print('\n Coeficiente de correlación')
coef_Correl=np.sqrt(coef_Deter)
display(coef_Correl)
```

```python
pr1=mex[mex['room_type'] == 'Private room']
display(sns.scatterplot(x='host_response_rate', y='host_acceptance_rate', color="blue", data=pr1))
Vars_Indep= pr1[['host_response_rate']]
Var_Dep= pr1['host_acceptance_rate']
model= LinearRegression()
type(model)
model.fit(X=Vars_Indep, y=Var_Dep)
print('\n Predicción de modelo')
display(model.__dict__)
print('\n Eficiencia del modelo')
display(model.score(Vars_Indep,Var_Dep))
print('\n Predecimos valores')
y_pred= model.predict(X=pr1[['host_response_rate']])
display(y_pred)
print('\n Agregamos predicciones')
pr1.insert(0, 'Predicciones', y_pred)
display(pr1)
print('\n Coeficiente de determinación')
coef_Deter=model.score(X=Vars_Indep, y=Var_Dep)
display(coef_Deter)
print('\n Coeficiente de correlación')
coef_Correl=np.sqrt(coef_Deter)
display(coef_Correl)
```
